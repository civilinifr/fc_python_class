We will download earthquake data using tools developed by IRIS (Incorporated Research Institutions for Seismology):

https://www.iris.edu/hq/

They maintain and curate data from seismic stations all around the world. There are many different ways to get data from them, but I'm personally a fan of the routine *PyWeed*, and that's what we are going to use today:

https://ds.iris.edu/ds/nodes/dmc/software/downloads/pyweed/

We can use the utility to select seismic stations and the earthquake data (or **events**) recorded at those stations.

<img src="https://raw.githubusercontent.com/civilinifr/fc_deeplearning/master/eq_detector_images/station_data_location.png" alt="Drawing" style="width: 1200px;">

In this case, we are downloading all of the earthquakes magnitude 3 and above that are within 1 degree distance (approximately 110 km) from a site called PFO (Pinon Flat Observatory) in California. **Location** is a number designating the instrument at a particular site (sites may have multiple instruments), and **channel** is an IRIS code that specifies instrument information. 

In short, the first latter refers to the samplerate of the instrument (how many data points it records per second), the second to the type of instrument (certain types of seismometers are better at recording nearby earthquakes while others are more suited for distant earthquakes), and the last to the directional component being recored (most seismometers will record motion across two horizontal directions and the vertical). We will pick the channel HHZ, which refers to a (H) high-samplerate (100 samples per second) (H) strong-motion accelerometer (best resolution for nearby strong earthquakes) recording in the (Z) vertical direction. Once you've selected all the earthquakes, you can download the traces. 

An earthquake is composed of the following types of waves (in order): pressure (P-wave), shear (S-wave), and surface (Rayleigh and Love). In this exercise, we are interested in identifying the start of the earthquake as quickly as possible. The IRIS dataset contains P-wave arrivals (onset of the P-wave at the seismometer) for each earthquake. In order to get noise prior to the earthquake arrival, we pick our data traces to span 101 seconds before to 60 seconds past the P-wave arrival:


<img src="https://raw.githubusercontent.com/civilinifr/fc_deeplearning/master/eq_detector_images/selected_seismograms.png" alt="Drawing" style="width: 1200px;">


As you can see from the output list, some of the earthquakes don't record any earthquake data (3.4 Ml 2005-08-31) and others have an incorrect P-wave arrival time (4.0 Ml 2005-08-31). Make sure to go through the earthquakes and remove those types of events from the waveform preview prior to download. For output file type, choose miniseed (SAC is probably fine too, but the file sizes tend to be a bit bigger).

Before we go on, a quick note about time-keeping in seismology. It's kind of a mess because everyone seems to have different conventions. In our case, since we're building our own images in matplotlib, we can get a matplotlib friendly output using the following: 

```python
tr.times("matplotlib")
```

However, we don't really care so much about knowing the absolute time and we're much better off using relative time. The matplotlib time format is days, so we can do a conversion from days to seconds (24 x 60 x 60 = 86400) after subtracting the start element from the entire vector.

```python
time = (tr.times("matplotlib")-tr.times("matplotlib")[0])*86400
```

We can then get the element in the time vector that corresponds to the P-wave arrival (i.e., everything before is noise and everything after is the earthquake). 

```python
# What time sample corresponds to start of the P-wave? 
# We know it's 101 seconds after, so we can use it to find the appropriate index
P_ind = int((tr.stats.sampling_rate * 101) + 1)
```

```python
# Plot a figure with the P-wave arrival
fig = plt.figure(figsize =(10,5))
ax = fig.add_subplot(1, 1, 1)
ax.plot(time, tr.data, "b-")
fig.autofmt_xdate()
plt.plot([time[P_ind], time[P_ind]], [np.nanmin(tr.data), np.nanmax(tr.data)], c='black')

plt.ylabel('Amplitude (counts)', fontweight = 'bold')
plt.xlabel('Time (s)', fontweight = 'bold')
plt.show()
```

That's a nice looking seismogram! 

Hmmm.... we can see here that the P-wave arrival isn't quite exactly right. Unfortunately there is very little we can do about it because it's based on the P-wave arrival registered for the file. We should be able to build a machine-learning model to find appropriate P-wave arrival times using spectograms.... However, that's a project for another time :-) 

We will NOT normalize the amplitudes of the seismograms like we did for the music instrument recognizer. This is because we do not want to add a computationally intensive operation since we'd like this to detect the arrival as soon as possible. 

In order to make our data augmentation work (more on that in a little bit), we need to be able to "cut" a seismogram and its resulting spectogram at different times. We accomplish this using the functions plot_seismo and plot_spectro. As in the music instrument recognizer, the plot_spectro routine has a vmax value to prevent outlier data from skewing the color scheme that is largely flexible. I tried a few different ones and `vmax=250000` seemed to strike a good balance for showing earthquake intensities for various magnitudes. 

**NOTE:** In this particular spectogram function, I used `imshow` to plot the spectogram. It looks real pretty, but I found that it produces timing errors during certain intervals of the data augmentation. This example is the only time I use imshow for plotting the spectogram and will use `pcolormesh` for the remainder of the code. 

```python
# Create a function that will plot the seismograph for a given time window

def plot_seismo(input_trace, input_start, input_end):
    
    # Convert the time of the input trace from matplotlib to seconds
    time = (input_trace.times("matplotlib")-input_trace.times("matplotlib")[0])*86400
    
    # Convert the input time to indices
    input_start_ind = int((input_trace.stats.sampling_rate * input_start))
    input_end_ind = int((input_trace.stats.sampling_rate * input_end))
    
    # Plot the seismogram
    fig = plt.figure(figsize =(10,5))
    ax = fig.add_subplot(1, 1, 1)
    ax.plot(time, input_trace.data, "b-")
    plt.ylabel('Amplitude (counts)', fontweight = 'bold')
    plt.xlabel('Time (s)', fontweight = 'bold')
    plt.xlim((time[input_start_ind],time[input_end_ind]))

    plt.show()
    
    return

# Plot the trace between 95-125 seconds
plot_seismo(tr, 95, 125)
```

```python
# Plot a spectogram using the scipy signal module we used for the instrument recognizer

def plot_spectro(input_trace, input_start, input_end):
    
    # Convert the time of the input trace from matplotlib to seconds
    time = (input_trace.times("matplotlib")-input_trace.times("matplotlib")[0])*86400
    
    # Convert the input time to indices
    input_start_ind = int((input_trace.stats.sampling_rate * input_start))
    input_end_ind = int((input_trace.stats.sampling_rate * input_end))
    
    # Cut the data to increase speed
    time_cut = time[input_start_ind:input_end_ind]
    data_cut = input_trace.data[input_start_ind:input_end_ind]

    time_axis = np.arange(len(data_cut))/input_trace.stats.sampling_rate
    f, t, Sxx = signal.spectrogram(np.array(data), input_trace.stats.sampling_rate)
    
    # Add the first element of the time_cut so we don't start from time = 0 
    t = t+time_cut[0]
    
    # Plot the spectogram
    plt.figure(figsize=(10,5))
    plt.imshow(Sxx, cmap=cm.jet, interpolation='bilinear', vmax=250000, origin='lower', aspect='auto',
               extent=(min(t),max(t),min(f),max(f)))
    plt.ylabel('Frequency (Hz)', fontweight='bold')
    plt.xlabel('Time (sec)', fontweight='bold')
    plt.xlim((time[input_start_ind],time[input_end_ind]))
    plt.colorbar() 
    
    return Sxx

# Plot the spectogram of the trace between 95-125 seconds.
Sxx = plot_spectro(tr, 95, 125)
```

# Explore data augmentation

We now need to run data augmentation on all of the files. To keep things clear, we will create a few additional folders: a train folder which will have earthquake and noise spectograms for the training set, a valid folder which will have earthquake and noise spectograms for the validation set, and seismograms from each of the 4 possible combinations in case we need to check errors on certain spectograms.

```python
path = '/media/francesco/Elements/ML_projects/data/'
! ls {path}
```

As previously stated, we will use 20 second slices within the noise and the earthquake portions of the signal with 1 second overlap for data augmentation. We begin at time 0 seconds for the noise and at time 81 seconds for the earthquake (19 seconds prior to the onset of the P-wave at 101 seconds). 

```python
noise_iterations = np.arange(0,20)
eq_iterations = np.arange(81,101)

noise_start, noise_end = [], []
eq_start, eq_end = [], []

for iteration in noise_iterations:
    noise_start.append(iteration)
    noise_end.append(iteration+20)
    
for iteration in eq_iterations:
    eq_start.append(iteration)
    eq_end.append(iteration+20)
    
```

We can use the function below to do data augmentation for one trace to see that it's working properly. 

```python
def plot_seismo_spectro_aug(input_trace, input_start, input_end, output_folder):
    
    # Convert the time of the input trace from matplotlib to seconds
    time = (input_trace.times("matplotlib")-input_trace.times("matplotlib")[0])*86400
    
    # Convert the input time to indices
    input_start_ind = int((input_trace.stats.sampling_rate * input_start))
    input_end_ind = int((input_trace.stats.sampling_rate * input_end))
    
    f, t, Sxx = signal.spectrogram(input_trace.data, input_trace.stats.sampling_rate)

    fig = plt.figure(figsize =(10,10))
    ax = fig.add_subplot(2, 1, 1)
    ax.plot(time, input_trace.data, "b-")
    plt.ylabel('Amplitude (counts)', fontweight = 'bold')
    plt.xlabel('Time (s)', fontweight = 'bold')
    plt.xlim((time[input_start_ind],time[input_end_ind]))
    
    ax2 = fig.add_subplot(2, 1, 2)
    specmax = 250000
    plt.pcolormesh(t, f, Sxx, cmap=cm.jet, vmax=specmax)
    plt.ylabel('Frequency (Hz)', fontweight='bold')
    plt.xlabel('Time (sec)', fontweight='bold')
    
    plt.xlim(input_start, input_end)
    
    # Designate a good filename
    filename = (f'seismo_spectra_{input_trace.stats.network}_{input_trace.stats.station}_'
                f'{input_trace.stats.channel}_{input_trace.stats.starttime}_'
                f'{input_start}_{input_end}.png')

    plt.savefig(f'{output_folder}{filename}')
    
    plt.close(fig)
    
    return
    
```

```python
# Run the data augmentation for an example earthquake
example_file = f'{path}/data_valid/AZ.PFO..HHZ_2008-06-24T15_04_32.869_2008-06-24T15_06_43.869.mseed'
iterations = np.arange(len(noise_start))

for iteration in iterations:
    st = read(example_file)
    tr = st[0]
    plot_seismo_spectro_aug(tr, eq_start[iteration], eq_end[iteration], f'{path}aug_test/eq/')

print(f'Saved file {os.path.basename(example_file)}...')
```

## Apply the data augmentation

We now are going to apply the data augmentation for all files in the training and validation sets (and remove things like axis and labels). The codes are going to start getting longer, but most of the routines are very similar. We will be using the Python `parallel` library to speed up the processing. We are saving figures, so the total file spaced produced should only be a few hundred megabytes at most. 

```python
train_files = glob.glob(f'{path}data_train/*.mseed')
valid_files = glob.glob(f'{path}data_valid/*.mseed')
```

```python
noise_iterations = np.arange(0,20)
eq_iterations = np.arange(81,101)

noise_start, noise_end = [], []
eq_start, eq_end = [], []

for iteration in noise_iterations:
    noise_start.append(iteration)
    noise_end.append(iteration+20)
    
for iteration in eq_iterations:
    eq_start.append(iteration)
    eq_end.append(iteration+20)
    
```

```python
# A more augmentation friendly version of the plot seismo function
def plot_seismo_aug(input_trace, input_start, input_end, output_folder):
    
    # Designate a good filename
    filename = (f'seismo_{input_trace.stats.network}_{input_trace.stats.station}_'
                f'{input_trace.stats.channel}_{input_trace.stats.starttime}_'
                f'{input_start}_{input_end}.png')
    
    if not os.path.exists(f'{output_folder}{filename}'):
    
        # Convert the time of the input trace from matplotlib to seconds
        time = (input_trace.times("matplotlib")-input_trace.times("matplotlib")[0])*86400

        # Convert the input time to indices
        # We will use the Obspy stats variables so we can apply this to all types of seismometers
        input_start_ind = int((input_trace.stats.sampling_rate * input_start))
        input_end_ind = int((input_trace.stats.sampling_rate * input_end))

        # Plot the seismogram
        fig = plt.figure(figsize =(10,5))
        ax = fig.add_subplot(1, 1, 1)
        ax.plot(time, input_trace.data, "b-")
        plt.xlim((time[input_start_ind],time[input_end_ind]))

        mpl.rcParams['savefig.pad_inches'] = 0
        plt.axis('off')

        ax = plt.axes([0,0,1,1], frameon=False)
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
        plt.autoscale(tight=True)
        plt.autoscale(tight=True)

        plt.savefig(f'{output_folder}{filename}', bbox_inches='tight')

        plt.close(fig)
        
    else:
        return
    
    return
```

```python
# Plot a spectogram using the scipy signal module we used for the instrument recognizer

def plot_spectro_aug(input_trace, input_start, input_end, output_folder):
    
    filename = (f'spectra_{input_trace.stats.network}_{input_trace.stats.station}_'
                f'{input_trace.stats.channel}_{input_trace.stats.starttime}_'
                f'{input_start}_{input_end}.png')
    
    if not os.path.exists(f'{output_folder}{filename}'):
    
        # Convert the time of the input trace from matplotlib to seconds
        time = (input_trace.times("matplotlib")-input_trace.times("matplotlib")[0])*86400

        # Find the input start of the 
        input_start_ind = int((input_trace.stats.sampling_rate * input_start))
        input_end_ind = int((input_trace.stats.sampling_rate * input_end))

        f, t, Sxx = signal.spectrogram(input_trace.data, input_trace.stats.sampling_rate)

        specmax = 250000

        # Plot the spectogram
        fig = plt.figure(figsize=(10,5))
        plt.pcolormesh(t, f, Sxx, cmap=cm.jet, vmax=specmax)
        plt.xlim(input_start, input_end)

        mpl.rcParams['savefig.pad_inches'] = 0
        plt.axis('off')
        ax = plt.axes([0,0,1,1], frameon=False)
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
        plt.autoscale(tight=True)
        plt.autoscale(tight=True)

        # Designate a good filename
        plt.savefig(f'{output_folder}{filename}', bbox_inches='tight')

        plt.close(fig)
        
    else:
        return
    
    return
    
```

```python
%%capture

# Run the time and spectral domain data augmentation on the training set

def run_augmentation(input_file): 
    
    iterations = np.arange(len(noise_start))
    for iteration in iterations:
        st = read(input_file)
        tr = st[0]
        plot_seismo_aug(tr, noise_start[iteration], noise_end[iteration], f'{path}seismo_train_noise/')
        plot_seismo_aug(tr, eq_start[iteration], eq_end[iteration], f'{path}seismo_train_eq/')
        plot_spectro_aug(tr, noise_start[iteration], noise_end[iteration], f'{path}train/noise/')
        plot_spectro_aug(tr, eq_start[iteration], eq_end[iteration], f'{path}train/earthquake/')

    print(f'Saved file {os.path.basename(input_file)}...')
    
    return

num_cores = 7
Parallel(n_jobs=num_cores)(delayed(run_augmentation)(train_file) for train_file in train_files)
```

```python
%%capture
# Run the time and spectral domain data augmentation on the validation set
def run_augmentation(input_file): 
    
    iterations = np.arange(len(noise_start))
    for iteration in iterations:
        st = read(input_file)
        tr = st[0]
        plot_seismo_aug(tr, noise_start[iteration], noise_end[iteration], f'{path}seismo_valid_noise/')
        plot_seismo_aug(tr, eq_start[iteration], eq_end[iteration], f'{path}seismo_valid_eq/')
        plot_spectro_aug(tr, noise_start[iteration], noise_end[iteration], f'{path}valid/noise/')
        plot_spectro_aug(tr, eq_start[iteration], eq_end[iteration], f'{path}valid/earthquake/')

    print(f'Saved file {os.path.basename(input_file)}...')
    
    return

num_cores = 7
Parallel(n_jobs=num_cores)(delayed(run_augmentation)(valid_file) for valid_file in valid_files)
```

```python
# Compute the probability by applying the model

def compute_prob(input_trace, f, t, Sxx, input_start, input_end):
    
    if not os.path.exists(f'{runtime_path}tmp/'):
        os.mkdir(f'{runtime_path}tmp/')
    
    # Save the figure to the test folder
    filename = (f'spectra_{input_trace.stats.network}_{input_trace.stats.station}_'
                f'{input_trace.stats.channel}_{input_trace.stats.starttime}_'
                f'{input_start}_{input_end}.png')

    specmax = 250000

    fig = plt.figure(figsize=(10,5))
    plt.pcolormesh(t, f, Sxx, cmap=cm.jet, vmax=specmax)
    plt.xlim(input_start, input_end)

    mpl.rcParams['savefig.pad_inches'] = 0
    plt.axis('off')
    ax = plt.axes([0,0,1,1], frameon=False)
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    plt.autoscale(tight=True)
    plt.autoscale(tight=True)

    plt.savefig(f'{runtime_path}tmp/{filename}', bbox_inches='tight')
    plt.close(fig)
    
    # Test the figure
    img = open_image(f'{runtime_path}tmp/{filename}')
    classify, pred_idx, outputs = learn.predict(img)
            
    return classify, pred_idx, outputs
```

```python
# Modified augmentation code and probability code wrapper

def test_model(input_trace, input_start, input_end):
    
    
    # Convert the time of the input trace from matplotlib to seconds
    time = (input_trace.times("matplotlib")-input_trace.times("matplotlib")[0])*86400
    
    # Convert the input time to indices
    # We will use the Obspy stats variables so we can apply this to all types of seismometers
    input_start_ind = int((input_trace.stats.sampling_rate * input_start))
    input_end_ind = int((input_trace.stats.sampling_rate * input_end))
    
    # Compute probability
    f, t, Sxx = signal.spectrogram(input_trace.data, input_trace.stats.sampling_rate)
    classify, idx, outputs = compute_prob(input_trace, f, t, Sxx, input_start, input_end)
    probability = f'{np.round(outputs[idx].data.item()*100,2)}%'
    
    if idx==0:
        classification = 'earthquake'
    else:
        classification = 'noise'
    
    # Plot and save result
    fig = plt.figure(figsize=(12,7))
    
    ax1 = plt.subplot2grid((2,3), (0,0), colspan=2, rowspan=1)
    ax1.plot(time, input_trace.data, "b-")
    if idx == 0:
        ax1.plot([input_start, input_end, input_end, input_start, input_start],
                 [np.min(input_trace.data),np.min(input_trace.data), np.max(input_trace.data), np.max(input_trace.data), np.min(input_trace.data)], 
                 c='red')
    else:
        ax1.plot([input_start, input_end, input_end, input_start, input_start],
                 [np.min(input_trace.data),np.min(input_trace.data), np.max(input_trace.data), np.max(input_trace.data), np.min(input_trace.data)], 
                 c='gray')

    plt.ylabel('Amplitude (counts)', fontweight = 'bold')
    plt.xlabel('Time (s)', fontweight = 'bold')
    
    output_str = f'Prediction: {classification} - {probability} confidence'    
    plt.title(output_str)
    
    ax2 = plt.subplot2grid((2,3), (0,2), colspan=1, rowspan=1)
    ax2.plot(time, input_trace.data, "b-")
    plt.ylabel('Amplitude (counts)', fontweight = 'bold')
    plt.xlabel('Time (s)', fontweight = 'bold')
    plt.xlim((time[input_start_ind],time[input_end_ind]))
    
    ax3 = plt.subplot2grid((2,3), (1,2), colspan=1, rowspan=1)
    specmax = 250000
    ax3.pcolormesh(t, f, Sxx, cmap=cm.jet, vmax=specmax)
    plt.ylabel('Frequency (Hz)', fontweight='bold')
    plt.xlabel('Time (sec)', fontweight='bold')
    plt.xlim((time[input_start_ind],time[input_end_ind]))    
    
    ax4 = plt.subplot2grid((2,3), (1,0), colspan=1, rowspan=1)
    ax4.axis('off')
    ax4.get_xaxis().set_visible(False)
    ax4.get_yaxis().set_visible(False)
    
    if idx==0:
        plt.text(0.5, 0.5, output_str, fontweight='bold', bbox=dict(facecolor='red', alpha=0.5), 
                 verticalalignment='center', transform=ax4.transAxes, size=15)
    else:
        plt.text(0.5, 0.5, output_str, fontweight='bold', bbox=dict(facecolor='gray', alpha=0.5), 
                 verticalalignment='center', transform=ax4.transAxes, size=15)        

    fig.tight_layout()
    
    plt.subplots_adjust(wspace=0.5)
    
    filename = (f'model_{input_trace.stats.network}_{input_trace.stats.station}_'
                f'{input_trace.stats.channel}_{input_trace.stats.starttime}_'
                f'{input_start}_{input_end}.png')
    
    
    plt.savefig(f'{runtime_path}{filename}', bbox_inches='tight')
    
#     plt.show()
    
    plt.close(fig)
    
    return

total_seconds = int(tr.stats.npts / tr.stats.sampling_rate)
start_times = np.arange(0,total_seconds-20)
end_times = np.arange(20,total_seconds) 

for iteration in np.arange(len(start_times)):
    test_model(tr, start_times[iteration], end_times[iteration])
```
